## Data pre processing

```python
path = '../processeddata/individual_stocks_5yr/individual_stocks_5yr/'
ticker = 'AAPL' # Change Stock to analyze

p = path + ticker + '_data.csv' 
df = pd.read_csv(p)

df.set_index(['date'],inplace=True)
df = df[['close']]

# Days to Forecast
forecast_out = 10

df['Prediction'] = df['close'].shift(-forecast_out)
df.dropna(inplace=True)

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

X = np.array(df.drop(['Prediction'], 1))
X = X[:-forecast_out]


Y = np.array(df['Prediction'])
Y = Y[:-forecast_out]

X_train_full, X_test, Y_train_full, Y_test = train_test_split(X, Y, test_size = 0.5, shuffle = False)
X_train, X_valid, Y_train, Y_valid = train_test_split(X_train_full, Y_train_full, shuffle = False)

X_train = X_train / 200
Y_train = Y_train / 200
X_test = X_test /200
Y_test = Y_test /200

X_valid = X_valid / 200
Y_valid = Y_valid / 200
```

## Part A

```python

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(100, activation="relu"))
model.add(tf.keras.layers.Dense(1, activation="relu"))


model.compile(optimizer="adam", loss= "mse")


model.fit(X_train, Y_train, epochs=100, validation_data=(X_valid, Y_valid))

print(model.evaluate(X_test, Y_test))
```

## Part C

```python
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)
```

```python
def build_model(n_hidden=2, n_neurons=100, learning_rate=3e-3, input_shape=[1]):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation="relu"))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.SGD(lr=learning_rate)
    model.compile(loss="mse", optimizer=optimizer)
    return model
```

```python
keras_reg = keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

```python
keras_reg.fit(X_train, Y_train, epochs=100,
              validation_data=(X_valid, Y_valid),
              callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```

```python
mse_test = keras_reg.score(X_test, Y_test)
```

```python
y_pred = keras_reg.predict(X_test)
```

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
from scipy.stats import reciprocal
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {
    "n_hidden": [0, 1, 2, 3],
    "n_neurons": np.arange(1, 100),
    "learning_rate": reciprocal(3e-4, 3e-2),
}

rnd_search_cv = RandomizedSearchCV(keras_reg, param_distribs, n_iter=10, cv=3, verbose=2)
rnd_search_cv.fit(X_train, Y_train, epochs=100,
                  validation_data=(X_valid, Y_valid),
                  callbacks=[keras.callbacks.EarlyStopping(patience=10)])
```
