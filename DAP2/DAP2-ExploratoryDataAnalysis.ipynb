```python
path = '../processeddata/SP500_Prices.csv'
df = pd.read_csv(path)

df.set_index(['Date'],inplace=True)

display(df)
print(df.describe())

print("\n Mean:")
print(df.mean())
 
print("\n Median:")
print(df.median())
 
print("\n Mode:")
print(df.mode())
```

# all_stocks_5yr.csv

```python
#Data Frame Descibe
path = '../processeddata/all_stocks_5yr.csv'
df = pd.read_csv(path)

print(display(df))
print("Description:")
print(df.describe())
```

```python
# We are only selecting a company from each sector

# Amazon (Consumer Discretionary), Apple (IT),  JPMorgan Chase & Co. (Financial), PACCAR Inc. (Industrial), 
# Aetna Inc (Health Care), Costco Wholesale Corp. (Consumer Staples), Simon Property Group (Real Estate), Exxon Mobil Corp. (Energy)
# PG&E Corp. (Utilities), Air Products & Chemicals Inc (Materials), AT&T Inc. (Telecommunication Services) 
path = '../processeddata/individual_stocks_5yr/individual_stocks_5yr/'

comp = ['AMZN', 'AAPL', 'JPM', 'PCAR', 'AET', 'COST', 'SPG', 'XOM', 'PCG', 'APD', 'T']

for c in comp:
    p = path + c + '_data.csv' 
    df = pd.read_csv(p)
    df.set_index(['date'],inplace=True)

    lines = df['close'].plot.line()
```

```python
# Comparing Amazon to Nordstrom
compare = ['JWN', 'AMZN']

for c in compare:
    p = path + c + '_data.csv' 
    df = pd.read_csv(p)
    df.set_index(['date'],inplace=True)

    df['close'].plot.line()
```
